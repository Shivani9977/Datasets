```python
def check_gaussian(data):
    sns.histplot(data,kde=True)
    plt.show()
    stats.probplot(data,dist='norm',plot=plt)
    plt.show()
    a=stats.anderson(data,dist='norm')
    if (a.statistic>a.critical_values[2]):
        print('column is non-gassian')
    else:
        print('column is gaussian')
```

```python
def fix_outliers_NG(data):
    q1=data.quantile(0.25)
    q3=data.quantile(0.75)
    iqr=q3-q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    data = np.where(data < lower_bound, lower_bound,
            np.where(data > upper_bound, upper_bound, data))
    return data
    
```

```python
df['NumberOfPersonVisiting']=fix_outliers_NG(df['NumberOfPersonVisiting'])
```

```python
model=[
    KNeighborsClassifier(),
    LogisticRegression(),
    RidgeClassifier(),
    RandomForestClassifier(),
    DecisionTreeClassifier(),
    SVC(),
    
]
```

```python
trained_model=[]
for i in model:
    xnew,ynew=x_train,y_train
    m=i.fit(xnew,ynew)
    trained_model.append(m)
    
```

```python
for model in trained_model:  
    y_pred = model.predict(x_test)
    acc = accuracy_score(y_test, y_pred)
    print(f"{model.__class__.__name__} Accuracy: {acc}")
```
